# ⁃	Length:
⁃	Q4: You will want to create a new column with the integer value of the number of pages.(If you remove rows in this process, please state why.) 

I add a new column called "pages_num" as the integer value of the number of pages. Since NaN cannot be transferred to int, I use -1 instead of it to represent NaN.

```python
pages_num = df['book_pages'].copy()
pages_num = pages_num.str.replace(" pages",'').fillna(-1).astype(int)
df['pages_num'] = pages_num
df
```

⁃	Q5: What is the average number of pages?

```python
df[df['pages_num']>=0]['pages_num'].mean()
```

⁃	Q6: What is the median number of pages? 

```python
df[df['pages_num']>=0]['pages_num'].median()
```

⁃	Q7: Does having more than 1 author result in a longer book on average? What is the average number of pages for books written by a single author? What is it for when there are at least two authors? (We will do more careful analysis for these types of questions later. For now, we just want you to practice using some DataFrame functionalities). 

```python
df_single = df.loc[~df['book_authors'].str.contains('\|')]
df_single[df_single['pages_num']>=0]['pages_num'].mean()
```

```python
df_multiple = df.loc[df['book_authors'].str.contains('\|')]
df_multiple[df_multiple['pages_num']>=0]['pages_num'].mean()
```

The average number of pages for books written by a single author is 334.2275641025641. And the average number of pages for books written by at least two authors is 284.02962962962965. So having more than 1 author doesn't result in a longer book on average.

# Ratings
⁃	Q8: How many books have at least a rating of 4.25? 

```python
df[df['book_rating'] >= 4.25].shape[0]
```

⁃	Q9: How about at most a rating of 4?

```python
df[df['book_rating'] <= 4].shape[0]
```

⁃	Q10: Discretize the ratings. (1,2,3,4,5). 

```python
pd.cut(x = df['book_rating'], bins = [1,2,3,4,5])
```

⁃	Q11: For each of the discretized ratings (1,2,3,4,5), what is the average number of pages?

```python
df_countablepages = df[df['pages_num'] >= 0]
df_countablepages.groupby(pd.cut(x = df_countablepages['book_rating'], bins = [1,2,3,4,5]))['pages_num'].mean()
```

⁃	Q12: How about the average number of reviews?

```python
df.groupby(pd.cut(x = df['book_rating'], bins = [1,2,3,4,5]))['book_review_count'].mean()
```

# Genres:
⁃	Q13: Create a new DataFrame, exploding the rows with multiple genres such that it is one row per genre/book.

```python
genres = df['genres'].str.split('|',expand = True).stack()
genres = genres.reset_index(level=1,drop=True).rename('genre')
df2 = df.copy()
df2 = df2.join(genres)
df2
```

⁃	Q14: What is the average rating of books from different genres? What is the median?

```python
df2.groupby(df2['genre'])['book_rating'].mean()
```

```python
df2.groupby(df2['genre'])['book_rating'].median()
```

⁃	Q15: What is the average length of books from different genres? What is the median?

```python
df2_countablepages = df2[df2['pages_num']>=0]
df2_countablepages.groupby(df2_countablepages['genre'])['pages_num'].mean()
```

```python
df2_countablepages.groupby(df2_countablepages['genre'])['pages_num'].median()
```
